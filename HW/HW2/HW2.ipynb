```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('study_performance.csv')
df
```

Q1:每一組學生有無完成考試準備課程對數學成績的影響

```python
sns.violinplot(data=df, x="race_ethnicity", y="math_score",hue="test_preparation_course")
```

A1:有完成考試準備課程學生的成績中位數都略高於同一組未完成課程學生的成績，且低於20分的人數明顯較少，另外在group A中也可以發現超過100分的學生中大部分是有完成考試準備課程的學生。

Q2:父母學歷對於學生數學成績的影響

```python
sns.violinplot(data=df, x="parental_level_of_education", y="math_score")
```

A2:在父母不同學歷的學生數學成績中，中位數最高的是master's degree，而high school和some high school則略低於其他四者且成績分布差距較大，master's degree相較於其他而言資料較為集中且沒有出現分數低於20分的學生，而some college整體略低於bachelor's degree但分布情形類似。
