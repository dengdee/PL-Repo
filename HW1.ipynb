```python
import pandas as pd

#2023-12-05羽球男單世界排名
df1 = pd.read_csv("2023-12-05 BWF world ranking man's singles.csv")

#2024-03-12羽球男單世界排名
df2 = pd.read_csv("2024-03-12 BWF world ranking man's singles.csv")

display(df1)
display(df2)

set1=set(df1['name'])
set2=set(df2['name'])
```

聯集
(輸出結果為曾經出現在前25名的所有人名)

```python
U = set1 | set2
print("Union:", U)
```

交集
(輸出結果為兩次都有出現在前25名的人名)

```python
I = set1 & set2 
print("Intersetion:", I)
```

差集
(輸出結果為兩次排名中只出現一次)

```python
#落榜
D2023 = set1 - set2

#新上榜
D2024 = set2 - set1

print("Difference(2023-12-05有上榜但2024-03-12沒有):", D2023)
print("Difference(2024-03-12有上榜但2023-12-05沒有):", D2024)
```
